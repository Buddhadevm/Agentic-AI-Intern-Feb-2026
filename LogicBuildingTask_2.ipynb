### Problem 1: Unique Words in a Sentence

```python
# Input sentence
sentence = input("Enter the sentence: ")

# Split into words (assuming spaces separate them)
words = sentence.split()

# Remove duplicates
unique_words = set(words)

# Print results
print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

### Problem 2: Highest Salary from Employee Data

```python
# Real world input system

employees = {}

while True:
    name = input("Enter employee name (type 'done' to stop): ")

    if name.lower() == "done":
        break

    salary = int(input("Enter salary: "))
    employees[name] = salary

employee_name = None
highest_salary = 0

for name, sal in employees.items():
    if sal > highest_salary:
        highest_salary = sal
        employee_name = name

print("Highest Salary:", employee_name, "-", highest_salary)
```

```python
# create dictionary
employees = {}

n = int(input("How many employees? "))

for i in range(n):
    name = input("Enter employee name: ")
    salary = int(input("Enter salary: "))
    employees[name] = salary

# variables
employee_name = None
highest_salary = 0

for name, sal in employees.items():
    if sal > highest_salary:
        highest_salary = sal
        employee_name = name

print("Highest Salary:", employee_name, "-", highest_salary)
```

### Problem 3: Find Maximum and Minimum Values

```python
# Real world input system

numbers = []

while True:
    val = input("Enter a number (type 'done' to finish): ")

    if val.lower() == "done":
        break

    numbers.append(int(val))

max_val = numbers[0]
min_val = numbers[0]

for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("Numbers:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

```python
numbers = []

n = int(input("How many numbers? "))

for i in range(n):
    value = int(input("Enter number: "))
    numbers.append(value)

# assume first number is both max and min
max_val = numbers[0]
min_val = numbers[0]

for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

### Problem 4: Count Products Above a Price Threshold

```python
product_price = []

n = int(input("How many numbers? "))

for i in range(n):
    value = int(input("Enter number: "))
    product_price.append(value)

count = 0

for p in product_price:
    if p > 1000:
        count += 1

print("Products above 1000:", count)
```

### Problem 5: Calculate Attendance Percentage

```python
attendance = []

n = int(input("Enter number of days: "))

for i in range(n):
    status = input("Enter P for Present or A for Absent: ")
    attendance.append(status)

present_days = attendance.count("P")
total_days = len(attendance)

percentage = (present_days / total_days) * 100

print("Attendance list:", attendance)
print("Attendance Percentage:", percentage)
```

### Problem 6: Remove Duplicate Phone Numbers

```python
# with country code and exact 10 digit number

phone_numbers = []

n = int(input("How many phone numbers to enter? "))

for i in range(n):
    code = input("Enter country code (example +91): ")

    
    while True:
        number = input("Enter 10-digit phone number: ")

        if len(number) == 10 and number.isdigit():
            break
        else:
            print("Invalid input — phone number must be exactly 10 digits.")

    full_number = code + number
    phone_numbers.append(full_number)

# remove duplicates
unique_numbers = list(set(phone_numbers))

print("\nAll entered numbers:", phone_numbers)
print("Unique phone numbers:", unique_numbers)
```

### Problem 7: Count Character Frequency

```python
text = input("Enter a string: ")

freq = {}

for i  in text:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1

print("Character frequency:", freq)
```

### Problem 8: Convert List to Tuple

```python
values = []

n = int(input("How many elements in list? "))

for i in range(n):
    val = int(input("Enter value: "))
    values.append(val)

tuple = tuple(values)

print("List:", values)
print("Tuple:", tuple)
```

### Problem 9: Check if a Key Exists in a Dictionary

```python

data = {}

n = int(input("How many entries? "))

for i in range(n):
    name = input("Enter name: ")
    value = input("Enter value: ")
    data[name] = value

check_key = input("Enter name to search: ")

if check_key in data:
    print("Employee exists")
else:
    print("Key does not exist")
```

### Problem 10: Calculate Average Marks

```python
marks = []

n = int(input("How many subjects? "))

for i in range(n):
    m = float(input("Enter marks: "))
    marks.append(m)

total = sum(marks)
average = total / len(marks)

print("Marks list:", marks)
print("Average Marks:", average)
```
