## 1. Difference between "Love" and "love" in NLP

In NLP, "Love" and "love" are treated as different words if case normalization is not applied.
"Love" may be considered a proper noun or start of a sentence, while "love" is a general word.
Usually, we convert everything to lowercase to treat them as the same word.

## 2. What happens if stopwords are not removed?

If stopwords are not removed, the dataset will contain many unnecessary words like "is", "the", "and".
This increases noise and can reduce model performance by focusing on less meaningful words.

## 3. Two real-world scenarios where removing stopwords can be harmful

(a) Sentiment Analysis: Words like "not" are important (e.g., "not good" vs "good")
(b) Question Answering: Words like "who", "what", "where" are important for meaning


## 4. Difference between stemming and lemmatization

 Stemming: Cuts words to root form without considering meaning (e.g., "running" → "run")
Lemmatization: Converts words to meaningful base form using vocabulary (e.g., "better" → "good")

## Task 2:Build Advanced Preprocessing Function 

```python
import re

def preprocess_text(text):
    if not text or text.strip() == "":
        return []

    # Remove URLs and emails
    text = re.sub(r"http\S+|www\S+|@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Convert to lowercase
    text = text.lower()

    # Handle repeated characters (soooo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove special characters (keep only words)
    text = re.sub(r"[^a-z\s]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    tokens = text.split()

    # Remove short tokens (≤2 except 'no', 'not')
    cleaned_tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    return cleaned_tokens
```

## Task 3: Stress Testing 

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    cleaned_sentence = " ".join(tokens)

    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned_sentence)
    print("-" * 50)
```

## Task 4: Token Analytics 

```python
def token_analytics(tokens):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(word) for word in tokens) / total_tokens if total_tokens > 0 else 0

    return total_tokens, unique_tokens, avg_length

for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    total, unique, avg = token_analytics(tokens)

    print("Sentence:", sentence)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Length:", avg)
    print("-" * 50)
```

## Task 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []

for sentence in test_sentences:
    all_tokens.extend(preprocess_text(sentence))

counter = Counter(all_tokens)

print("Top 10 most frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[-5:])
```

## Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(" ".join(tokens))

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

result = full_pipeline(test_sentences)
print(result)
```
