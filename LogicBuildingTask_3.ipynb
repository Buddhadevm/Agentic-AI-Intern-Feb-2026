### Problem Statement 1: Employee Performance Bonus Eligibility 

```python

employees = {}
n = int(input("Enter number of employees: "))

for _ in range(n):
    name = input("Enter employee name: ").strip()
    score = int(input(f"Enter performance score for {name}: "))
    employees[name] = score

# Find highest score
max_score = max(employees.values())

# Find all employees having that score (tie handling)
top_performers = [name for name, score in employees.items() if score == max_score]

print("\nTop Performers Eligible for Bonus:")
print(", ".join(top_performers), f"(Score: {max_score})")
```

### Problem Statement 2: Search Query Keyword Analysis 

```python

import string

query = input("Enter customer search query: ")

# Convert to lowercase
query = query.lower()

# Remove punctuation (real-world messy input handling)
for ch in string.punctuation:
    query = query.replace(ch, "")

words = query.split()

freq = {}

for word in words:
    freq[word] = freq.get(word, 0) + 1

# Show only repeated keywords
repeated = {k: v for k, v in freq.items() if v > 1}

print("\nRepeated keywords found:")
print(repeated)
```

### Problem Statement 3: Sensor Data Validation

```python

raw = input("Enter hourly sensor readings separated by space: ")
sensor_readings = list(map(int, raw.split()))

valid_readings = []

# Even readings = valid
for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((hour, value))

print("\nValid Sensor Readings (Hour, Value):")
print(valid_readings)
```

### Problem Statement 4: Email Domain Usage Analysis 

```python
emails = []
n = int(input("Enter number of employee emails: "))

for _ in range(n):
    emails.append(input("Enter email: ").strip().lower())

# Dictionary to store domain → count
domain_count = {}

#  Extract domain from each email and count usage
for email in emails:
    domain = email.split("@")[-1]
    domain_count[domain] = domain_count.get(domain, 0) + 1  # Increase count safely using get()

total = len(emails)

print("\nDomain Usage Percentage:")

# Convert counts into percentage usage
for domain, count in domain_count.items():
    percent = (count / total) * 100
    print(f"{domain}: {percent:.0f}%")
```

### Problem Statement 5: Sales Spike Detection

```python

raw = input("Enter daily sales values separated by space: ")
sales = list(map(int, raw.split()))

# Average Sales
avg_sales = sum(sales) / len(sales)

# To Find Days where Sales are more than 30% above average
threshold = avg_sales * 1.30

print(f"\nAverage Sales: {avg_sales:.2f}")
print("Sales Spike Days:")

for day, value in enumerate(sales, start=1):
    if value >= threshold:
        print(f"Day {day}: {value}")
```

### Problem Statement 6: Duplicate User ID Detection

```python

raw = input("Enter user IDs separated by space: ")
user_ids = raw.split()

# Dictionary to store user_id → frequency
count_map = {}

# Count each occurrence
for uid in user_ids:    # If ID already exists, increase count, Otherwise start from 1
    count_map[uid] = count_map.get(uid, 0) + 1

print("\nDuplicate User IDs Found:")

# Show only IDs that appear more than once
for uid, count in count_map.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
