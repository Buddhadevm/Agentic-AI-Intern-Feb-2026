### Problem 1: Smart Light Controller 

```python
import speech_recognition as sr

# Initialize smart home with predefined rooms and Each room light starts in OFF state.
class SmartHomeLights:
    
    def __init__(self):

        self.lights = {
            "bedroom": "OFF",
            "studyroom": "OFF",
            "hall": "OFF",
            "sitting room": "OFF",
            "kitchen": "OFF",
            "bathroom": "OFF"
        }
    
    #  This function processes the spoken command.

    def process_voice_command(self, command):
     
        command = command.lower()

        room_found = None
        action_found = None

        # Detect room name
        for room in self.lights:
            if room in command:
                room_found = room
                break

        # Detect action
        if "on" in command:
            action_found = "ON"
        elif "off" in command:
            action_found = "OFF"

        # Validation
        if room_found is None or action_found is None:
            print("Command not recognized.")
            return

        # Update light status
        self.lights[room_found] = action_found

        print(f"{room_found.title()} light is {action_found}")
    
 # Capture voice input from microphone and convert speech to text   
def listen_command():

    recognizer = sr.Recognizer()

    with sr.Microphone() as source:
        print("Speak your command...")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)

    try:
        command = recognizer.recognize_google(audio)
        print("You said:", command)
        return command

    except:
        print("Voice not recognized.")
        return ""


home = SmartHomeLights()

# Capture voice
voice_command = listen_command()

# Process command
home.process_voice_command(voice_command)
```

### Problem 2: Employee ID Card System 

```python
class EmployeeID:
    
    def __init__(self, name, emp_id, department):
        # Store employee details
        self.name = name
        self.emp_id = emp_id
        self.department = department
    
    def display_card(self):
        
        print("\nEmployee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Taking input
name = input("Enter employee name: ")
emp_id = input("Enter employee ID: ")
dept = input("Enter department: ")

employee1 = EmployeeID(name, emp_id, dept)
employee1.display_card()
```

### Problem 3: Mobile Contact Record 

```python
class Contact:
    
    def __init__(self, name, phone):
        # Store contact details
        self.name = name
        self.phone = phone
    
    def display_contact(self):
       
        print("\nContact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Taking input
name = input("Enter contact name: ")
phone = input("Enter phone number: ")

contact1 = Contact(name, phone)
contact1.display_contact()
```

### Problem 4: Product Price Tag Generator

```python
class Product:
    
    def __init__(self, product_name, price):
        # Store product details
        self.product_name = product_name
        self.price = price
    
    def display_price_tag(self):
        
        print("\nProduct:", self.product_name)
        print(f"Price: ₹{self.price}")


# Taking input
product = input("Enter product name: ")
price = float(input("Enter product price: "))

item1 = Product(product, price)
item1.display_price_tag()
```

### Problem 5: Movie Rating Display System

```python
class Movie:
    
    def __init__(self, movie_name, rating):
        # Store movie information
        self.movie_name = movie_name
        self.rating = rating
    
    def display_movie(self):
        
        print("\nMovie:", self.movie_name)
        print(f"Rating: {self.rating} / 5")


# Taking input
movie = input("Enter movie name: ")
rating = float(input("Enter movie rating: "))

movie1 = Movie(movie, rating)
movie1.display_movie()
```
