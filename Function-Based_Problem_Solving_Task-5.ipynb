### Problem 1 — Social Media Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
    
    total_likes = 0
    
    # Calculate total using loop
    for like in likes_list:
        total_likes += like
    
    print("Total Likes:", total_likes)
    
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Taking input from user
likes = list(map(int, input("Enter likes separated by space: ").split()))
analyze_engagement(likes)
```

### Problem 3 — Agriculture Rainfall Adequacy Checker

```python
def analyze_rainfall(rainfall_data, required_level):
    
    total = 0
    
    for rain in rainfall_data:
        total += rain
    
    average = total / len(rainfall_data)
    
    print("Average Rainfall:", average)
    
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


rainfall = list(map(int, input("Enter seasonal rainfall values: ").split()))
required = float(input("Enter required rainfall level: "))
analyze_rainfall(rainfall, required)
```

### Problem 4 — Duplicate Account Detection

```python
def detect_duplicates(usernames):
   
    unique_users = set()
    duplicate_found = False
    
    for user in usernames:
        if user in unique_users:
            duplicate_found = True
            break
        unique_users.add(user)
    
    if duplicate_found:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


users = input("Enter usernames separated by space: ").split()
detect_duplicates(users)
```

### Problem 6 — Premium Crop Price Filter

```python
def filter_premium_crops(price_list):
    
    premium = []
    
    for price in price_list:
        if price > 2000:
            premium.append(price)
    
    print("Premium Crops:", premium)


prices = list(map(int, input("Enter crop prices separated by space: ").split()))
filter_premium_crops(prices)
```

### Problem 9 — E-Learning Attendance Eligibility

```python
def check_exam_eligibility(attendance_list):
   
    present = 0
    
    for status in attendance_list:
        if status.upper() == "P":
            present += 1
    
    percentage = (present / len(attendance_list)) * 100
    
    print("Attendance Percentage:", percentage)
    
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


attendance = input("Enter attendance record (P/A separated by space): ").split()
check_exam_eligibility(attendance)
```

### # Problem 10: Smart Electricity Bill Analyzer

```python
def calculate_bill(units):
   
    bill = 0
    
    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    
    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    
    return bill, status


units_input = int(input("Enter electricity units consumed: "))
total_bill, usage_status = calculate_bill(units_input)

print("Total Bill:", total_bill)
print("Usage Classification:", usage_status)
```
